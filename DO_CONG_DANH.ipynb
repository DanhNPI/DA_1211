### 1. Đọc và tìm hiểu dữ liệu

```python
import pandas as pd

df = pd.read_csv('./bank_transactions.csv', encoding='utf-8')
df.head()
```

```python
df.info()
```

### 2. Thông tin GD trong ngày giữa 2 NH có lượng tiền TB trong mỗi giao dịch max, min

### 3. NH có số lần giao dịch gửi tiền nhiều nhất trong ngày, trong tháng

```python
df1 = df.groupby(['MONTH','DAY','FROM_CODE'])['TRANS'].aggregate('sum').unstack()
df1
```

```python
df1.max().sort_values(ascending=False)
```

#### ==> NH BIDV có số lần giao dịch gửi tiền nhiều nhất trong ngày, trong tháng

### 4. NH có số lần giao dịch nhận tiền nhiều nhất trong ngày, trong tháng

```python
df2 = df.groupby(['MONTH','DAY','TO_CODE'])['TRANS'].aggregate('sum').unstack()
df2
```

```python
df2.max().sort_values(ascending=False)
```

#### ==> NH MSB có số lần giao dịch gửi tiền nhiều nhất trong ngày, trong tháng

### 5. Tìm cặp NH có số lần GD qua lại nhiều nhất

```python
df.head()
```

```python
df3 = df.groupby(['FROM_CODE','TO_CODE'])['TRANS'].sum().sort_values(ascending=False)
df3
```

#### ==> Cặp NH BIDV - MSB có số lần giao dịch qua lại nhiều nhất

### 6. Tổng số GD trong từng tháng và trong từng ngày của các NH

### 8. Số GD gửi tiền đi trong từng ngày của NH MB

```python
df7 = df.loc[df['FROM_CODE'] == 'MB']
df7.groupby('DAY')['TRANS'].count()
```

```python
# NH MB chỉ thực hiện GD gửi tiền đi trong 2 ngày là ngày mùng 2 và ngày mùng 10
# Số lần giao dịch của mỗi ngày là tương đối ít
```

### 9. Số GD nhận tiền đi trong từng ngày của NH MB

```python
df8 = df.loc[df['TO_CODE'] == 'MB']
df8
```

```python
df8.groupby('DAY')['TRANS'].count()
```

```python
# NH MB thực hiện GD nhận tiền trong 4 ngày 2,6,9,10
# Trong đó ngày 10 có số GD nhận tiền là lớn nhất
```
